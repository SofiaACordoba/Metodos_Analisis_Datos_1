```python
#------------------Aisgnamos rango de edad segun DNI-----------------------------------------

def asignar_rango_etario(dni):
    dni = int(dni)
    if dni < 5_000_000:
        return '90-99 años'
    elif dni < 8_000_000:
        return '80-89 años'
    elif dni < 12_000_000:
        return '70-79 años'
    elif dni < 18_000_000:
        return '60-69 años'
    elif dni < 25_000_000:
        return '50-59 años'
    elif dni < 32_000_000:
        return '40-49 años'
    elif dni < 39_000_000:
        return '30-39 años'
    elif dni < 46_000_000:
        return '20-29 años'
    elif dni < 49_000_000:
        return '17-19 años'
    else:
        return 'Extranjeros'


df['RANGO_ETARIO'] = df['DNI'].apply(asignar_rango_etario)

print(df[['DNI', 'RANGO_ETARIO']].head())
print("\nDistribución por rango etario:") 
print(df['RANGO_ETARIO'].value_counts())


extranjeros = df[df['RANGO_ETARIO'] == 'Extranjeros']
print(extranjeros['DNI'])


```

```python


# Filter for Aprobado only
df_aprobados = df[df['NOTA'] == 'Aprobado']

# Group by COMISION and count Aprobado
resumen_comision = df_aprobados.groupby('COMISION').size()

# Print summary
print("\nNúmero de Aprobados por Comisión:")
print(resumen_comision)

# Vertical bar chart for number of Aprobado
plt.figure(figsize=(12, 8))  
resumen_comision.plot(kind='bar', color='#4CAF50')  # Green bars
plt.title('Número de Aprobados por Comisión', fontsize=16)
plt.xlabel('Comisión', fontsize=14)
plt.ylabel('Cantidad de Aprobados', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate labels for readability
plt.tight_layout(pad=2.0)  # Extra padding
plt.xticks(rotation=90, ha='center', fontsize=8)  # Vertical labels, smaller font
plt.subplots_adjust(bottom=0.25)  # More space for labels
plt.show()

# Duplicates analysis (for completeness)
duplicados = df_aprobados[df_aprobados.duplicated(subset=['DNI', 'NRO. INSCR'], keep=False)]
resumen_duplicados = duplicados.groupby('COMISION').size()
print("\nNúmero de Aprobados en filas duplicadas por Comisión:")
print(resumen_duplicados)
```
